```python
import seaborn as sns
import matplotlib.pyplot as plt
import  numpy as np
import pandas as pd
%matplotlib inline
```

```python
#### Data Cleaning
```

```python
df[df.isnull().any(axis = 1)]
```

```python
df.loc[:122,"Region"] = 0
df.loc[122:,"Region"] = 1
```

```python
df.info()
```

```python
df['Region'] = df['Region'].astype(int)
```

```python
df.head(),df.tail()
```

```python
df.isnull().sum()
```

```python
## remove null
```

```python
df = df.dropna().reset_index(drop= True)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.iloc[[122]]
```

```python
df = df.drop(122).reset_index(drop=True)
```

```python
df.iloc[[122]]
```

```python
df.columns = df.columns.str.strip()
```

```python
df.info()
```

```python
df[['day','month','year','Temperature','RH','Ws']]=df[['day','month','year','Temperature','RH','Ws']].astype(int)
```

```python
objects = [features for features in df.columns if df[features].dtypes=='O']
```

```python
for i in objects:
    if i!= 'Classes':
        df[i] = df[i].astype(float)
```

```python
df.info()
```

```python
df.describe()
```

```python
df.to_csv('Algerian_Cleaned_dataset2.csv')
```
